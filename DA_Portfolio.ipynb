```python
# load the file and duplicate

customer_csv = pd.read_csv("customer_shopping_behavior.csv")
customer_csv.tail()
data = customer_csv.copy()
data
```

```python
# get the summary of num and cat using the include parameter

data.describe(include="all")
```

```python
# sum of null values

data.isnull().sum()
```

```python
# impute missing values in the data using median based on each categories

data["Review Rating"] = data.groupby("Category")["Review Rating"].transform(lambda x : x.fillna(x.median()))
data["Review Rating"].head()
```

```python
# recheck for null vals

data.isnull().sum()
```

```python
# reformat cols into lower case - for easy reference of code

data.columns = data.columns.str.lower()
data.columns
```

```python
# replace spaces with underscore

data.columns = data.columns.str.replace(" ", "_")
data.columns
```

```python
# map and rename

data = data.rename(columns= {"purchase_amount_(usd)" : "purchase_amount"})
data.columns
```

```python
# feature engineering to create new column for age 

labels = ["young adult", "adult", "middle-aged", "senior"]
data["age_grp"] = pd.qcut(data["age"], q=4, labels = labels)
data[["age", "age_grp"]].head(10)
```

```python
# purchase frequency engineering

freq_map = {"Fortnightly" : 14,
            "Weekly" : 7,
            "Monthly" : 30,
            "Quarterly" : 90,
            "Bi-Weekly" : 14,
            "Annually" : 365,
            "Every 3 Months" : 90}

data["freq_purch"] =data["frequency_of_purchases"].map(freq_map)
data[["frequency_of_purchases", "freq_purch"]].head(10)
```

```python
# remove col 

data = data.drop("promo_code_used", axis=1)
```
